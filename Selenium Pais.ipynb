# Scraping all the cities names from the autocomplete search bar

```python

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

# Defining the path for the driver
PATH = "C:\Program Files (x86)\chromedriver.exe"

# Creating the driver
driver = webdriver.Chrome(PATH)

#Getting the wanted website
driver.get('https://www.pais.co.il/PointsOfSale/')

#Printinh websites title
print(driver.title)

# Creating two empty lists to hold cities names
cities = []
cities_unique = []

#Creating alphabet list
my_str = 'אבגדהוזחטיכלמנסעפצקרשת'

def split(word):
    return list(word)

alphaBet = split(my_str)


#Running through the alphaBet
for letter in alphaBet:
    
    #Finding the search bar
    search_input = driver.find_element_by_id("node")
    search_return = driver.find_element_by_id('mainSearchBtn')
    
    #Filling the search bar with a letter of the alphabet
    search_input.send_keys(letter)
    
    #Pausing for 5 seconds so the code can run smoothly
    time.sleep(3)
    #search_return.send_keys(Keys.RETURN)
    
    
    # Sraching for the element that holds the city name
    l=driver.find_elements_by_class_name("ui-menu-item")
    
    #appending the name of the city to 'cities list'
    for i in l:
        print(i.text)
        cities.append(i.text)
    
    #Clearing the search bar in order to add another letter
    driver.find_element_by_id('node').clear()

# Getting the unique values for cities    
for city in cities:
    if city not in cities_unique:
        cities_unique.append(city)
        
```

# Will extract all the adresses of lottery stations

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Path to the driver
PATH = "C:\Program Files (x86)\chromedriver.exe"


#Setting the driver
driver = webdriver.Chrome(PATH)

#Getting the wanted website
driver.get('https://www.pais.co.il/PointsOfSale/')
print(driver.title)

#Finding the search bar and search button
search_input = driver.find_element_by_id("node")
search_return = driver.find_element_by_id('mainSearchBtn')


#Creating empty lists for the addresses and cities
adress_list = []
city_list = []

# Filling the search bar with the unique cities names and hitting RETUrn on the search button
for city in cities_unique:
    search_input = driver.find_element_by_id("node")
    search_return = driver.find_element_by_id('mainSearchBtn')
    search_input.send_keys(city)
    search_return.send_keys(Keys.RETURN)
    time.sleep(3)
    
    #As long as True, try to see if there is a load more button, if so,press it.
    while True:
        try:
        #identify element
            if city == 'בני עי"ש':
                break
            d = driver.find_element_by_id("loadmore_items")
            s= d.text
            print("Element exist -" + s)
            d.click()
            time.sleep(3)
        #NoSuchElementException thrown if not present
        except Exception as e:
            print (e)
            break

    #Creating a list of all elements in the webpage that are addresses        
    l=driver.find_elements_by_class_name("adress_txt")
    
    #Appending the addresses and their cities to two lists
    for i in l:
        print(city)
        print(i.text)
        city_list.append(city)
        adress_list.append(i.text)
```

```python
#Creating a dataframe with cities and addresses
df_pais = pd.DataFrame({'Cities':city_list,'Address':adress_list})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df_pais['Cities'].value_counts().head(5)
```

```python
#Dataframe of cities populations from wikipedia
df_pop = pd.read_excel('cities_population_wiki.xlsx',sheet_name='Sheet2')
```

```python
#Merging the addresses with the population table
df_merged = pd.merge(df_pais,df_pop, on='Cities',how = 'inner')
```

```python
#Creating a column in a new data frame of the unique cities.
df_cities = pd.DataFrame({'Cities':df_merged['Cities'].unique()})
```

```python
#Merging the unique sities with their population
df_cities = pd.merge(df_cities,df_pop,on = 'Cities',how = 'inner')
```

```python
df_cities.head(5)
```

```python
#New dataframe with the count of their values
df_freq = pd.DataFrame(df_pais['Cities'].value_counts())
```

```python
#Resetting the index
df_freq = df_freq.reset_index()
```

```python
#Renaming the columns
df_freq = df_freq.rename(columns = {'index':'Cities','Cities':'lottery_num'})
```

```python
#Merging the cities with frequency of their appearence which is the number of lottery station in each city.
df_cities = pd.merge(df_cities,df_freq,on = 'Cities',how = 'inner')
```

```python
# Creating a new column - "How many people per station"
df_cities['population/station'] = df_cities['Population']/df_cities['lottery_num']
```
